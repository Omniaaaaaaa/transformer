```python
import numpy as np

def softmax(x):
    x = x - np.max(x, axis=-1, keepdims=True)
    e_x = np.exp(x / 1.5)   # temperature to make attention smoother
    return e_x / e_x.sum(axis=-1, keepdims=True)

def normalize(x):
    return x / (np.linalg.norm(x, axis=-1, keepdims=True) + 1e-8)

def self_attention(Q, K, V):
    d_k = Q.shape[-1]

    # Normalize Q and K to avoid very sharp attention values
    Q = normalize(Q)
    K = normalize(K)

    # Attention scores = QK^T / sqrt(d_k)
    scores = np.dot(Q, K.T) / np.sqrt(d_k)

    # Convert scores into probabilities
    attention_weights = softmax(scores)

    # Final attention output
    output = np.dot(attention_weights, V)

    return scores, attention_weights, output


# Sentence tokens
tokens = ['What', 'are', 'the', 'symptoms', 'of', 'diabetes', '?']

np.random.seed(42)

seq_len = len(tokens)
d_model = 8
d_k = 4
d_v = 4

# Random embeddings and weight matrices
X = np.random.randn(seq_len, d_model) * 0.5
W_Q = np.random.randn(d_model, d_k) * 0.5
W_K = np.random.randn(d_model, d_k) * 0.5
W_V = np.random.randn(d_model, d_v) * 0.5

# Compute Q, K, V
Q = X @ W_Q
K = X @ W_K
V = X @ W_V

# Apply self-attention
scores, weights, output = self_attention(Q, K, V)

print("Tokens:")
print(tokens)

print("\nAttention Scores:")
print(np.round(scores, 3))

print("\nAttention Weight Matrix:")
print(np.round(weights, 3))

print("\nHow much each token attends to 'symptoms':")
for i, token in enumerate(tokens):
    print(f"{token:12s} -> {weights[i, 3]:.3f}")

print("\nOutput Representations:")
print(np.round(output, 3))
```
